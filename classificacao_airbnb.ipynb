### Mostra a frequência unívoca dos valores de cada variável
- Se a frequência for `1` esta variável será `retirada` da lista de `features` e não será considerada na análise

```python
def count_frequence_values(df):
    for col in df:
        if pd.api.types.is_float_dtype(df[col]): 
            resp = len(df[col].unique())
            if resp == 1:
                features.remove(col)
                reject_vars_list.append(col)
            else:
                print('float   -> {} = {}'.format(col, resp))
        elif pd.api.types.is_integer_dtype(df[col]): 
            resp = len(df[col].unique())
            if resp == 1:
                features.remove(col)
                reject_vars_list.append(col)
            else:
                print('integer -> {} = {}'.format(col, resp) )
        elif pd.api.types.is_object_dtype(df[col]): 
            resp = len(df[col].unique())
            if resp == 1:
                features.remove(col)
                reject_vars_list.append(col)
            else:
                print('object  -> {} = {}'.format(col, resp) )
        elif pd.api.types.is_bool_dtype(df[col]): 
            resp = len(df[col].unique())
            if resp == 1:
                features.remove(col)
                reject_vars_list.append(col)
            else:
                print('boolean -> {} = {}'.format(col, resp) )
        else:
            resp = len(df[col].unique())
            if resp == 1:
                features.remove(col)
                reject_vars_list.append(col)
            else:
                print('outros  -> {} = {}'.format(col, resp) )
    return  resp 

count_frequence_values(df[features])
print('\n --- Lista de variáveis rejeitadas ---\n', reject_vars_list)
```

#### Coloca para `category` as variáveis de tipo `object` e `boolean`

```python
def df_change_type(df, features):
    for atrib in df[features]:
        if pd.api.types.is_object_dtype(df[atrib]):
            df[atrib] = df[atrib].astype('category')
        if pd.api.types.is_bool_dtype(df[atrib]):
            df[atrib] = df[atrib].astype('category') 
    return df

df = df_change_type(df, features)
df.shape
```

### Selecionar variáveis com alta correlação e baixa informação mútua
- Um limite baseado em variância é uma das primeira etapa de seleção de variáveis (features)
- A variação pode ser usada como um filtro para identificar colunas a serem removidas do conjunto de dados. 
- Uma coluna com um único valor tem uma variação de 0.0, e uma coluna com muito poucos valores exclusivos pode ter uma pequena variação
- Objetivo
    - Reduzir a incerteza de uma variável
    - Correlação
        - Baixa: 0,10 a 0,29
        - Médio: 0,30 a 0,49
        - Alta: 0,50 a 1,00
    - Os intervalos também se aplicam a valores negativos entre 0 e -1
    - Colunas que têm uma variação menor que **`0.50`** serão removidas

- Referências:
    - https://eng.libretexts.org/Bookshelves/Industrial_and_Systems_Engineering/Book%3A_Chemical_Process_Dynamics_and_Controls_(Woolf)/13%3A_Statistics_and_Probability_Background/13.13%3A_Correlation_and_Mutual_Information
    - https://en.wikipedia.org/wiki/Mutual_information
    - https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.VarianceThreshold.html
    - https://stackabuse.com/applying-filter-methods-in-python-for-feature-selection

```python
print(X_train.shape, X_valid.shape)
print(y_train.groupby('room_type').size())
```

```python
from sklearn.feature_selection import VarianceThreshold

constant_filter = VarianceThreshold(threshold=0.55)
constant_filter.fit(X_train)
constant_cols = [col for col in X_train.columns
                if col not in X_train.columns[constant_filter.get_support()]]
print('\n----------- Variáveis removidas -----------\n %s' %constant_cols)

remaining_cols = [col for col in X_train.columns
                 if col in X_train.columns[constant_filter.get_support()]]
print('\nVariáveis restantes com variação acima 50%% -> %d' % len(X_train.columns[constant_filter.get_support()]))
print('\n----------- Variáveis restantes -----------\n %s' %remaining_cols)

X_train = constant_filter.transform(X_train)
X_valid = constant_filter.transform(X_valid)

X_train = pd.DataFrame(data=X_train, columns=remaining_cols)
X_valid = pd.DataFrame(data=X_valid, columns=remaining_cols)

X_remaining = constant_filter.transform(X_remaining)
X_remaining = pd.DataFrame(data=X_remaining, columns=remaining_cols)

X_train.shape, X_valid.shape, X_remaining.shape
```

#### Removendo features `correlacionadas` acima de 80%
- Ref: https://stackabuse.com/applying-filter-methods-in-python-for-feature-selection/

```python
corr_features = set()
corr_matrix = X_train.corr()
    
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i,j]) > 0.8:
            col = corr_matrix.columns[i]
            corr_features.add(col)
print('Total de variáveis com correlação acima de 80%% -> %d' %len(corr_features))
print('\n---------- Variáveis removidas ----------\n%s' %corr_features)

features = list(X_train.columns[1])
print('\n---------- Variáveis remanescentes ----------\n%s' %features)

X_train.drop(labels=corr_features, axis=1, inplace=True)
X_valid.drop(labels=corr_features, axis=1, inplace=True)
X_remaining.drop(labels=corr_features, axis=1, inplace=True)

X_train.shape, X_valid.shape, X_remaining.shape
```
